```python
from scipy import stats
# 假设我们关注的是第二年级的学生（假设年级用数字“2”表示）
specific_grade = '大四'

# 筛选出特定年龄的数据
df_specific_grade = 数据表[数据表['年龄'] == specific_grade]

# 创建会打多少分和的交互表婚姻状况
cross_tab_specific_grade = pd.crosstab(df_specific_grade['已婚'], df_specific_grade['婚姻状况'])

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

chi2_grade, p_grade, dof_grade, 
```
